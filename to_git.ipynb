```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./../csv/vgsales.csv')
df = pd.DataFrame(data)
#df
df.head()
```

```python
avg = df['Global_Sales'].mean()
```

```python
#Newly launched top 10 games

top_df = df[df.Global_Sales > avg]
# top_df = top_df.sort_values(by='Global_Sales', ascending = False)
df2 = top_df.sort_values(['Year','Global_Sales'], ascending = [False,False]).head(10)
df2[['Name','Year','Global_Sales']]
```

```python
#The genre of games where people play the  most.

grouped = df.groupby('Genre').sum()
grouped.sort_values('Global_Sales', ascending = False)
grouped.index[0]

```

```python
#Games which maintains the trends throughout the analysis period 

df2 = df[(df['Year']>=2015) & (df['Year']<=2020)]
df2.sort_values(by='Global_Sales', ascending = False).head(10)
```

```python
#The platform where most of the games are launching
grouped = df.groupby(['Platform']).size().sort_values(ascending = False).head(10)
grouped
```

```python
# Top 10 games which are popular in North America, Europe and japan
```

```python
df.groupby(['Genre'], sort = True).size()
# df2 = pd.DataFrame(grouped, columns = ['value'])
# df2.sort_values(by = 'value', ascending = False)

# df.groupby(['Name'], sort = False).sum() 
```

```python
#Which vendor publish more number of games.

grouped = df.groupby('Publisher').size()
df2 = pd.DataFrame(grouped, columns = ['value'])
df2.sort_values(by = 'value', ascending = False).head(10)
```

```python
#The years in which more and fewer numbers of games released.

grouped = df.groupby('Year').size()
df2 = pd.DataFrame(grouped, columns = ['value'])
top = df2.sort_values(by = 'value', ascending = False).head(1)
bottom = df2.sort_values(by = 'value', ascending = False).tail(1)
print(top,'\n',bottom)
```

```python
#Which vendor owns the highest market in each country category as well as in global sales.

grouped = df.groupby('Publisher').sum().reset_index()
g = grouped.sort_values('NA_Sales', ascending=False)
g[['Publisher','NA_Sales']].head(1)
```

```python
grouped = df.groupby('Publisher').sum().reset_index()
g = grouped.sort_values('EU_Sales', ascending=False)
g[['Publisher','EU_Sales']].head(1)
```

```python
grouped = df.groupby('Publisher').sum().reset_index()
g = grouped.sort_values('JP_Sales', ascending=False)
g[['Publisher','JP_Sales']].head(1)
```

```python
grouped = df.groupby('Publisher').sum().reset_index()
g = grouped.sort_values('Global_Sales', ascending=False)
g[['Publisher','Global_Sales']].head(1)
```

```python
gr = ['NA_Sales','JP_Sales','EU_Sales','Global_Sales']
for i in gr:
    grouped = df.groupby('Publisher').sum().reset_index()
    g = grouped.sort_values(i, ascending=False)
#     g['Publisher',i].head(1)
    print(g.head(1))
```

```python
# Widely accepted games in Japan and Europe only

grouped = df.groupby('Name').sum().reset_index()
g = grouped.sort_values('JP_Sales', ascending=False)
g2 = g.sort_values('EU_Sales', ascending=False)
g2[['Name','JP_Sales','EU_Sales']].head()
```

```python
#Top 10 games which popular in North America, Europe and japan 

grouped = df.groupby('Name').sum().reset_index()
g = grouped.sort_values('NA_Sales', ascending=False)
g[['Name','NA_Sales']].head(10)
```

```python
grouped = df.groupby('Name').sum().reset_index()
g = grouped.sort_values('EU_Sales', ascending=False)
g[['Name','EU_Sales']].head(10)
```

```python
grouped = df.groupby('Name').sum().reset_index()
g = grouped.sort_values('JP_Sales', ascending=False)
g[['Name','JP_Sales']].head(10)
```

```python
#Publisher of top 10 games in North-America and Europe

grouped = df.groupby('Publisher').sum().reset_index()
g = grouped.sort_values('NA_Sales', ascending=False)
g2 = g.sort_values('EU_Sales', ascending=False)
g2[['Publisher','NA_Sales','EU_Sales']].head(10)
```
